```python
import numpy as np


song_data = {}
amazon_id = {}
a = data_meta_2['songs'][np.where(data_meta_2['songs'].fillna(0))[0]].map(lambda x : [i['title'] for i in x])
ind_num = 0
for i in range(len(data_meta_2)):
    if i in a.index:
        for j in a[i]:
            song_data[ind_num] = j
            amazon_id[ind_num] = data_meta_2['amazon-id'][i]
            ind_num += 1
    else:
        song_data[ind_num] = 0
        amazon_id[ind_num] = data_meta_2['amazon-id'][i]
        ind_num += 1
```

```python
len(song_data)
```

```python
t = pd.DataFrame(list(amazon_id.values()))
d = pd.DataFrame(list(song_data.values()))
p = pd.concat((t, d), axis=1)
p.columns = ['amazon-id', 'songs']
data_meta_3 = pd.merge(p, data_meta_2, how='outer', on='amazon-id')
```

```python
data_meta_3.loc[data_meta_3["songs_x"] == "", "songs_x"] = data_meta_3[data_meta_3["songs_x"] == ""]["title"]
```
